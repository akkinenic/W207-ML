```python
def P2():

    # Initialize the CountVectorizer with the default parameters
    vectorizer = CountVectorizer()
    # Fit the vectorizer using train_data set
    vectorizer.fit(raw_documents=train_data)
    # transform the train data to get the sparse matrix 
    transformed_train_data = vectorizer.transform(raw_documents=train_data)
    
    #For item - a
    #The sparse matrix shape[1] would give the size of the vocabulary
    #nnz will return the count of non-zero values in a matrix
    
    print ('a. The size of the vocabulary is {0}.'.format(transformed_train_data.shape[1]))

    print ('   The average number of non-zero features per example is {0:.3f}.'.\
        format(transformed_train_data.nnz / float(len(train_data))))
    
    print ('   {0:.3f}% of the entries in the matrix are non-zero.'.\
        format(100*float(transformed_train_data.nnz) / (transformed_train_data.shape[0] *
                                                 transformed_train_data.shape[1])))
    
    
    #For item - b
    #the get_feature_names() will return all the features names unsorted
    
    feature_strings = vectorizer.get_feature_names()
    #sort the features
    feature_strings.sort()
    print ('\nb. The 0th  feature string is "{0}"'.format(feature_strings[0]))
    print ('   The last feature string is "{0}"'.format(feature_strings[-1]))
    
    
    #For item - c
    #We can pass custom vocabulary to the CountVectorizer function using 'vocabulary' parameter
    ownvocab = ["atheism", "graphics", "space", "religion"]
    vectorizer_ownvocab = CountVectorizer(vocabulary=ownvocab)
    
    #Fit and Transform the training data using custom vocabulary
    transformed_train_data_ownvocab = vectorizer_ownvocab.fit_transform(raw_documents=train_data)
    
    #This should give all the messages to use only the given vocabulary words as features
    print ('\nc. Dimensions of training vectors:', transformed_train_data_ownvocab.shape)
    

    print ('   The average number of non-zero features per example, when using a 4-word ' \
          'vocabulary, is {0:.3f}.'. format(transformed_train_data_ownvocab.nnz / float(len(train_data))))
    
    #For item - d
    # By default the countVectorizer would extract words that contains minimum two characters as features
    # The default ngram_range is (1,1) that extracts all the unigram words as features.
    # For bigram
    
    #Using analyzer word
    vectorizer_bigram_word = CountVectorizer(ngram_range=(2, 2), analyzer="word").fit(raw_documents=train_data)
    print('\nd. The size of the vocabulary using bigram (2,2) and "word"   analyzer is ',
          format(len(vectorizer_bigram_word.vocabulary_)))
    
    vectorizer_bigram_char = CountVectorizer(ngram_range=(2, 2), analyzer="char").fit(raw_documents=train_data)
    print('   The size of the vocabulary using bigram (2,2) and "char"   analyzer is ',
          format(len(vectorizer_bigram_char.vocabulary_)))
    #print((vectorizer_bigram_char.shape[1]))

    vectorizer_bigram_charwb = CountVectorizer(ngram_range=(2, 2), analyzer="char_wb").fit(raw_documents=train_data)
    print('   The size of the vocabulary using bigram (2,2) and "char_wb" analyzer is ',
          format(len(vectorizer_bigram_charwb.vocabulary_)))
    
    #Using analyzer word
    vectorizer_trigram_word = CountVectorizer(ngram_range=(2, 3), analyzer="word").fit(raw_documents=train_data)
    print('\n \n   The size of the vocabulary using trigram (2,3) and "word"   analyzer is ',
          format(len(vectorizer_trigram_word.vocabulary_)))
    
    vectorizer_trigram_char = CountVectorizer(ngram_range=(2, 3), analyzer="char").fit(raw_documents=train_data)
    print('   The size of the vocabulary using trigram (2,3) and "char"   analyzer is ',
          format(len(vectorizer_trigram_char.vocabulary_)))
    #print((vectorizer_bigram_char.shape[1]))

    vectorizer_trigram_charwb = CountVectorizer(ngram_range=(2, 3), analyzer="char_wb").fit(raw_documents=train_data)
    print('   The size of the vocabulary using trigram (2,3) and "char_wb" analyzer is ',
          format(len(vectorizer_trigram_charwb.vocabulary_)))
    
    
    # For item e
    
    vectorizer_pruned_min_10_vocab = CountVectorizer(min_df=10).fit(raw_documents=train_data)
    print('\ne. The size of the vocabulary after pruning words that appear in fewer than 10 documents ',
          format(len(vectorizer_pruned_min_10_vocab.vocabulary_)))
    
    
    # For item f
    
    vectorizer = CountVectorizer()
    print ('\nf. The fraction of the words in the dev data that are missing from the vocabulary is {0:.3f}% '.\
        format(100*float(len(set(vectorizer.fit(raw_documents=dev_data).get_feature_names()) -
                             set(vectorizer.fit(raw_documents=train_data).get_feature_names()))) /
               len(set(vectorizer.fit(raw_documents=dev_data).get_feature_names()))))
    
P2()
```

(3) Use the default CountVectorizer options and report the f1 score (use metrics.f1_score) for a k nearest neighbors classifier; find the optimal value for k. Also fit a Multinomial Naive Bayes model and find the optimal value for alpha. Finally, fit a logistic regression model and find the optimal value for the regularization strength C using l2 regularization. A few questions:

a. Why doesn't nearest neighbors work well for this problem?

b. Any ideas why logistic regression doesn't work as well as Naive Bayes?

c. Logistic regression estimates a weight vector for each class, which you can access with the coef\_ attribute. Output the sum of the squared weight values for each class for each setting of the C parameter. Briefly explain the relationship between the sum and the value of C.

[4 pts]

```python
def P3():

    #For KNN classifier, to find the optimal value of k
    #From the web, I found that the optimal value of k could lie anywhere between 1 to sqrt of sample size. 
    #so, I am taking k range from 1-60
    # The fitting would take time 
    
    #Initialize the CountVectorizer. Fit the train_data. And transform both train_data and dev_data
    vectorizer = CountVectorizer()
    vectorizer.fit(train_data)
    transformed_train_data = vectorizer.transform(train_data)
    transformed_dev_data = vectorizer.transform(dev_data)
    
    #Defining K Range
    k_range = list(range(1, 60))
    #creating a param dictionary to pass it to GridSearchCV and using f1_macro as scoring
    param_grid = dict(n_neighbors=k_range)

    neigh = KNeighborsClassifier()
    grid = GridSearchCV(neigh, param_grid, scoring='f1_macro')
    
    #Fit the grid using transformed training data
    grid.fit(transformed_train_data,train_labels)

    # Single best score achieved across all params (k)
    print ('\n\n K Nearest Neighbors Classifier')
    print ('\n  Single best score achieved across all params for (k)= {0} is {1:.3f}% '.\
                    format(grid.best_params_['n_neighbors'],grid.best_score_))
    
    
    #prediction on the dev data set
    neigh = KNeighborsClassifier(n_neighbors=grid.best_params_['n_neighbors'])
    neigh.fit(transformed_train_data,train_labels)
    dev_predicted_labels = neigh.predict(transformed_dev_data)
    print ('\n  The F1-Score for Predicting Dev Data for (k)= {0} is {1:.3f} '.\
        format(grid.best_params_['n_neighbors'],metrics.f1_score(y_true=dev_labels,
                                                                 y_pred=dev_predicted_labels,average="macro")))
    
    
    #For Multinomial Naive Bayes classifier, to find the optimal value of alpha
    
    #Defining alpha range
    alpha_range = list(np.arange(0.0,0.2,0.002))
    #creating a param dictionary to pass it to GridSearchCV and using f1_macro as scoring
    param_grid = dict(alpha=alpha_range)
    Multinomial_model = MultinomialNB()
    grid = GridSearchCV(Multinomial_model, param_grid, scoring='f1_macro')
    
    #Fit the grid using transformed training data
    grid.fit(transformed_train_data,train_labels)

    print ('\n\n Multinomial Naive Bayes Model')
    # Single best score achieved across all params (alpha)
    print ('\n  Single best score achieved across all params for (alpha)= {0} is {1:.3f}% '.\
                    format(grid.best_params_['alpha'],grid.best_score_))
   
    #prediction on the dev data set
    Multinomial_model = MultinomialNB(alpha=grid.best_params_['alpha'])
    Multinomial_model.fit(transformed_train_data,train_labels)
    dev_predicted_labels = Multinomial_model.predict(transformed_dev_data)
    print ('\n  The F1-Score for Predicting Dev Data for (alpha)= {0} is {1:.3f} '.\
        format(grid.best_params_['alpha'],metrics.f1_score(y_true=dev_labels,
                                                                 y_pred=dev_predicted_labels,average="macro")))
    
    #For Logistic Regression Model, to find the optimal value of C - regularization parameter
    #From the docs, the smaller value of C yields a greater impact on the output
    
    #Defining C Range
    c_range = list(np.arange(0.01,0.2,0.02))
    c_range = c_range + [1, 10, 100]
    #creating a param dictionary to pass it to GridSearchCV and using f1_macro as scoring
    param_grid = dict(C=c_range)
    lr_model = LogisticRegression(penalty='l2')
    grid = GridSearchCV(lr_model, param_grid, scoring='f1_macro')
    
    #Fit the grid using transformed training data
    grid.fit(transformed_train_data,train_labels)

    print ('\n\nLogistic Regression Model')
    # Single best score achieved across all params (alpha)
    print ('\n  Single best score achieved across all params for (C)= {0:.2f} is {1:.3f}% '.\
                    format(grid.best_params_['C'],grid.best_score_))
    
    #prediction on the dev data set
    lr_model = LogisticRegression(penalty='l2', C=grid.best_params_['C'])
    lr_model.fit(transformed_train_data,train_labels)
    dev_predicted_labels = lr_model.predict(transformed_dev_data)
    print ('\n  The F1-Score for Predicting Dev Data for (C)= {0:.2f} is {1:.3f} '.\
        format(grid.best_params_['C'],metrics.f1_score(y_true=dev_labels,
                                                                 y_pred=dev_predicted_labels,average="macro")))
   
    sum_of_squares = []
    avg_sum_of_squares = []
   
    #for each c
    for c in c_range:
        lr_model = LogisticRegression(penalty='l2', C=c)
        # fit the logisitc regression model using transformed train data
        lr_model.fit(transformed_train_data,train_labels)
        # calculate sum of squares for each label
        sum_of_squares_each_label=(list(np.square(lr_model.coef_).sum(axis=1)))
        #calculate average sum of squares of each label and add it to avg_sum_of_squares list
        avg_sum_of_squares.append(np.sum(sum_of_squares_each_label)/4)
        #pre-append c value
        sum_of_squares_each_label.insert(0,c)
        sum_of_squares.append(sum_of_squares_each_label)
           
    
    print('\n\n Sum of Squares Matrix for given C\n\n',
          pd.DataFrame(sum_of_squares, columns=['C','alt.atheism','comp.graphics','sci.space','talk.religion.misc']))
    
    plt.plot(c_range, avg_sum_of_squares)
    plt.xlabel("C")
    plt.ylabel("Average Sum of Squares")
    plt.title('C Vs Average Sum of Squares')
    
    return grid.best_params_['C']

best_C = P3()
```

**ANSWER**:

    a. Why doesn't nearest neighbors work well for this problem?
        
        The KNN is based on the assumption that the classification of an instance is more similar to the classification of 
        other instances that are near by in the vector space. Main compuatation here is to sort the documents in order to
        find the k nearest neighbors for the test document. 
        
        Also, KNN doesn't know which attributes are important. Like when computing distance between data points, KNN weighs 
        each attribute same as other attributes. So, common words that are frequent will have same influence on the 
        distance comapred to important attributes. 
        
        Also, if you have missing data, KNN doesn't perform very well. KNN also takes some time to train the model 
        compared to other classifiers.

    b. Any ideas why logistic regression doesn't work as well as Naive Bayes?
    
        Naive Bayes works well here because it handles missing data well. NB just exlcudes the attribute with missing data 
        when computing posterior probability. The assumption NB makes that attributes are conditionally independent to 
        each other class and so, it can peform well. 
        
        Naive Bayes can have problems if there is lot of correlation as it relies on marginal distribution of attributes
        for a given class to do classification. 
        
        In Logistic Regression, we want the output as boolean. It measures the relationship between a output variable Y, 
        here it is categorical and the independent variables are usually continuous by using probability scores as the 
        predicted values of the output variable.
        
        Also, given the number of less data points in the data set, the variance will be a problem for Logistic 
        Regression compared to Naive Bayes. so, NB has a higher bias but lower variance compared to logistic 
        regression given the size of the data set.

    c. Logistic regression estimates a weight vector for each class, which you can access with the coef_ attribute. Output the sum of the squared weight values for each class for each setting of the C parameter. Briefly explain the relationship between the sum and the value of C.
    
        Logistic Regression tries to choose the best parameters that gives the best fit to the data. This means 
        minimizing the error between what the model predicts Vs what the actual output is
    
        When we have small data set and lot of dependent variables, in our case it is 26,000+, the logistic regression 
        model will alter the parameters to fit the data more accurately. But, this performs poorly on the future data
        as there could be lot of new words. 
    
        Regularization is applying a penalty to increase the magnitude of parameter values in order to reduce overfitting.
        C is the inverse of the Regularization parameter 'λ' as C=1/λ. Larger λ means it is less likely that the parameters
        will be increased in magnitude to overfit the data. The graph depicts that the errors are minimized when we have 
        lower C value.
    

(4) Train a logistic regression model. Find the 5 features with the largest weights for each label -- 20 features in total. Create a table with 20 rows and 4 columns that shows the weight for each of these features for each of the labels. Create the table again with bigram features. Any surprising features in this table?

[5 pts]

```python

def Large_Weights_Table(vectorizer = CountVectorizer(),feature_space='default options'):
    vectorizer.fit(train_data)
    transformed_train_data = vectorizer.transform(train_data)
    transformed_dev_data = vectorizer.transform(dev_data)
    
    #Using Best C Value From the previous function
    lr_model = LogisticRegression(penalty='l2', C=best_C)
    
    #fit the model using transformed train data
    lr_model.fit(transformed_train_data,train_labels)

    #The 'result' list will hold all the features that last largest weights across each label
    result = []
    #For each class of the target variable
    for i in range(len(newsgroups_train.target_names)):
        #Get the coefficients for that class
        coef_each_label = lr_model.coef_[i]
        #np.argpartition will partition the entire list based on the specified number
        top_five = np.argpartition(coef_each_label, -5)[-5:]
        #Once we get the top five weights - sort the weights and get the indexes of those weights
        top_five_sorted=top_five[np.argsort(coef_each_label[top_five])]
        #create a row to hold weights across each label
        for index in top_five_sorted[::-1]:
            row = []
            row.append(vectorizer.get_feature_names()[index])
            for j in range(len(newsgroups_train.target_names)):
                row.append(lr_model.coef_[j][index])
            result.append(row)
    
    print('\n\nThe Features with the largest weights for each label using ', feature_space,'\n\n',
          pd.DataFrame(result, columns=['feature','alt.atheism','comp.graphics','sci.space','talk.religion.misc']))
    
def P4():
    Large_Weights_Table()
    Large_Weights_Table(CountVectorizer(ngram_range=(2, 2)), 'Bigram Features')
    
P4()
```

(5) Try to improve the logistic regression classifier by passing a custom preprocessor to CountVectorizer. The preprocessing function runs on the raw text, before it is split into words by the tokenizer. Your preprocessor should try to normalize the input in various ways to improve generalization. For example, try lowercasing everything, replacing sequences of numbers with a single token, removing various other non-letter characters, and shortening long words. If you're not already familiar with regular expressions for manipulating strings, see https://docs.python.org/2/library/re.html, and re.sub() in particular. With your new preprocessor, how much did you reduce the size of the dictionary?

For reference, I was able to improve dev F1 by 2 points.

```python
def empty_preprocessor(s):
    return s

def better_preprocessor(s):
    
    #convert all text into lower case
    s = s.lower()
    
    #Remove multiple white space
    #s = re.sub('\s+\s+', '', s)
    
    #Remove line break, tab, return
    #s = re.sub('[\n\t\r]+', '', s)
    
    #Remove quotes
    #s = re.sub('\&*[amp]*\;|gt+', '', s)
    
    #Remove tickers
    #s = re.sub('\$\w*', '', s)
    
    #Remove citations
    #s = re.sub('\@\w*', '', s)
   
    #Remove multiple punctuation marks
    #s = re.sub('[,@\'?\.$%_]', '', s)
    
    #Removing numbers and replacing it with a single token 'Number'
    s = re.sub('[\d+]', ' Number ', s)
     
    #Remove everything and keep only letters and numbers
    s = re.sub('/[^A-Za-z0-9 ]+/i', ' ', s)
  
    #Shorten Long words greater than 6
    s = re.sub("(\w{%d})\w+" % 6, "\\1", s)
    
    return s

def Apply_Regression(vectorizer=CountVectorizer(stop_words={'english','of','are','but','the',})):
    #Fit the vectorizer using train data
    vectorizer.fit(train_data)
    #Transform both train and dev data
    transformed_train_data = vectorizer.transform(train_data)
    transformed_dev_data = vectorizer.transform(dev_data)
    
    #Using Best C Value From the previous function
    lr_model = LogisticRegression(penalty='l2',C=best_C)
    
    #fit the model using transformed train data
    lr_model.fit(transformed_train_data,train_labels)
    
    #Predict on the transformed dev data 
    dev_predicted_labels = lr_model.predict(transformed_dev_data)
    
    return [metrics.f1_score(y_true=dev_labels,y_pred=dev_predicted_labels,average="macro"),len(vectorizer.vocabulary_)]
    
def P5():
    #Using stop words as english for better F1-Score
    vectorizer=CountVectorizer(preprocessor = empty_preprocessor,stop_words='english')
    [NoPre_f1_score,NoPre_vocab_size] = Apply_Regression(vectorizer)
    print ('\n  The F1-Score for Predicting Dev Data for (C)= {0:.2f} with NO pre-processors is {1:.2f}% '.\
        format(best_C,NoPre_f1_score*100))
   
    #Using preprocessor 
    vectorizer=CountVectorizer(preprocessor = better_preprocessor,stop_words='english')
    [pre_f1_score,pre_vocab_size] = Apply_Regression(vectorizer)

    print ('\n  The F1-Score for Predicting Dev Data for (C)= {0:.2f} with PRE-PROCESSOR is {1:.2f}% '.\
        format(best_C,pre_f1_score*100))
    
    print ('\n  The size of the dictionary is reduced by {0:.2f}% '.\
        format(((float)(NoPre_vocab_size - pre_vocab_size)/NoPre_vocab_size)*100))
    
P5()
```

(6) The idea of regularization is to avoid learning very large weights (which are likely to fit the training data, but not generalize well) by adding a penalty to the total size of the learned weights. That is, logistic regression seeks the set of weights that minimizes errors in the training data AND has a small size. The default regularization, L2, computes this size as the sum of the squared weights (see P3, above). L1 regularization computes this size as the sum of the absolute values of the weights. The result is that whereas L2 regularization makes all the weights relatively small, L1 regularization drives lots of the weights to 0, effectively removing unimportant features.

```python
def logistic_regression(penalty,C=best_C):
    #create vectorizer with default options
    vectorizer = CountVectorizer()
    #Fit the vectorizer using training data
    vectorizer.fit(train_data)
    #Transform both training data and dev data
    transformed_train_data = vectorizer.transform(train_data)
    transformed_dev_data = vectorizer.transform(dev_data)
    
    #create a regression model using l1 penalty using best_C and tol
    lr_model = LogisticRegression(penalty=penalty, C=C,tol=.01)
    #fit the model using transformed training data 
    lr_model.fit(transformed_train_data,train_labels)
    
    return lr_model.coef_


def P6():
    # Keep this random seed here to make comparison easier.
    np.random.seed(0)
    
    #Get the coefficient matrix when using l1 penalty and calculate the non-zero weights count
    l1_model_coef = logistic_regression(penalty='l1')
    l1_non_zero_weights_count = np.sum(np.apply_along_axis(np.count_nonzero, 1, l1_model_coef))
    
    #Get the coefficient matrix when using l2 penalty and calculate the non-zero weights count
    l2_model_coef = logistic_regression(penalty='l2')
    l2_non_zero_weights_count = np.sum(np.apply_along_axis(np.count_nonzero, 1, l2_model_coef))
    
    print('\n  The non zero weights count when used l1 penalty',l1_non_zero_weights_count)
    print('\n  The non zero weights count when used l2 penalty',l2_non_zero_weights_count)
    
    print ('\n  The non zero weights count is reduced by {0:.2f}% for L1 penalty compared to L2 penalty'.\
        format(((float)(l2_non_zero_weights_count - l1_non_zero_weights_count)/l2_non_zero_weights_count)*100))
    
    #we can also acheive this by sumper imposing on the data set itself as below
    #idx = (l1_model_coef).any(axis=0)
    #vectorizer = CountVectorizer()
    #vectorizer.fit(train_data)
    #transformed_train_data = vectorizer.transform(train_data)
    #transformed_dev_data = vectorizer.transform(dev_data)
    #new_trained_data = transformed_train_data[:,idx]
    #new_dev_data = transformed_dev_data[:,idx]
    
    vocab_size = []
    accuracy = []
    #Defining C Range
    c_range = list(np.arange(0.01,0.2,0.02))
    c_range = c_range + [1, 10, 100]
    
    for c in c_range:
        #Get the coefficient matrix for each c when using l1 penalty
        l1_model_coef = logistic_regression(penalty='l1', C=c)
        #Get the positions where any feature weight is non zero
        idx = (l1_model_coef).any(axis=0)
        
        #create a CountVectorizer with default options and fit the train data
        vectorizer = CountVectorizer()
        vectorizer.fit(train_data)
        
        #Get all the feature names
        vocab = vectorizer.get_feature_names()
        
        #Now using the indexes of the non-zero weights, create a new vocabulary
        new_vocab = [vocab[i] for i, x in enumerate(idx) if x]
        new_vectorizer = CountVectorizer(vocabulary=new_vocab)
        
        #Transform the training data and dev data using the new vocabulary vectorizer
        transformed_train_data = new_vectorizer.transform(train_data)
        transformed_dev_data = new_vectorizer.transform(dev_data)
        
        #Now create a L2 Penalty Logistic Regression model 
        lr_model = LogisticRegression(penalty='l2', C=c, tol=.01)
        # fit the logisitc regression model using transformed train data
        lr_model.fit(transformed_train_data,train_labels)
        #Predict on the dev data
        dev_predicted_labels = lr_model.predict(transformed_dev_data)
        
        vocab_size.append(len(new_vocab))
        accuracy.append(metrics.f1_score(y_true=dev_labels,y_pred=dev_predicted_labels,average="macro"))
      
    plt.figure(figsize=(14, 4))
    plt.subplot(1, 2, 1)
    plt.plot(vocab_size, accuracy)
    plt.xlabel("Vocabulary Size")
    plt.ylabel("Accuracy")
    plt.title('Vocabulary Size vs. Accuracy')
    
    plt.subplot(1, 2, 2)
    plt.plot(c_range, vocab_size)
    plt.xlabel("C")
    plt.ylabel("Vocabulary Size")
    plt.title('C vs. Vocabulary Size')
    
P6()
```

(7) Use the TfidfVectorizer -- how is this different from the CountVectorizer? Train a logistic regression model with C=100.

Make predictions on the dev data and show the top 3 documents where the ratio R is largest, where R is:

maximum predicted probability / predicted probability of the correct label

What kinds of mistakes is the model making? Suggest a way to address one particular issue that you see.

[4 pts]

```python
def P7():
    vectorizer = TfidfVectorizer()
    vectorizer.fit(train_data)
    transformed_train_data = vectorizer.transform(train_data)
    transformed_dev_data = vectorizer.transform(dev_data)
    
    lr_model = LogisticRegression(penalty='l2', C=100,tol=.01)
    lr_model.fit(transformed_train_data,train_labels)
    
    dev_predicted_labels = lr_model.predict(transformed_dev_data)
    print ('\n  The F1-Score for Predicting Dev Data for C=100 is {0:.3f} '.\
        format(metrics.f1_score(y_true=dev_labels, y_pred=dev_predicted_labels,average="macro")))
    dev_predic_probs=lr_model.predict_proba(transformed_dev_data)
    
    max_prob = dev_predic_probs.max(axis=1)
    
    Prob_correct_label = np.array([dev_predic_probs[i,j] for i,j in
                                   zip(range(len(dev_labels)),dev_labels)])
   
    R = max_prob / Prob_correct_label
   
    top_three = np.argpartition(R, -3)[-3:]
    #Once we get the top three R values - sort them and get the indexes of those records
    top_three_sorted=top_three[np.argsort(R[top_three])][::-1]
    for i in top_three_sorted:
        
        print('--------------------------------------------------------------------------------------\n\n')
        print('\n\n R-Score {0:.3f}'.format(R[i]))
        print('\n\n Predicted Label -->',newsgroups_train.target_names[dev_predicted_labels[i]])
        print('\n\n Actual Label -->',newsgroups_train.target_names[dev_labels[i]])
        print('\n \n Document ->\n ',dev_data[i])
    
P7()
```

**ANSWER:**
    
How is this different from the CountVectorizer?
    
    CountVectorizer gives equal weightage to all the words. i.e., each word is converted to a column and for each document, 
    it is equal to 1 if the word is present or 0 otherwise.
    
    TF-IDF will also has this information but will say how important that word is for that document with respect 
    to all the documents.
    
    As an example, the word 'The' has a higher frequency in most documents. But it doesn't give any good information for classification. So, TF-IDF will reduce the weightage of this word 'The' by taking log of number of documents divided by number of documents in which this term appears.
    
    then idf("the") = 0
    
    But if we have some word like 'computer' that appears only few times, we get idf as idf('computer') with some weight. 
    
    And the most infrequent words will get more weight. 
    
What kinds of mistakes is the model making? Suggest a way to address one particular issue that you see.
    
     From the results of the documents displayed above, we can see that other categories are wrongly classifed as graphics when they contain computer related terms such as 'internet/online/email' or 'printers'
     
     If we can somehow reduce the weights of such words or use such identified words as stop words, then we can reduce classification error.

```python
#Lets take a look at most important 10 features for each category when using tf-idf count vectorizer

def show_top10(classifier,vectorizer, categories):
    feature_names = np.asarray(vectorizer.get_feature_names())
    for i, category in enumerate(categories):
        top10 = np.argsort(classifier.coef_[i])[-10:]
        print("%s: %s" % (category, " ".join(feature_names[top10])))

def P8():
    vectorizer = TfidfVectorizer(sublinear_tf=True, ngram_range=(1, 2), stop_words=('english'))
    vectorizer.fit(train_data)
    transformed_train_data = vectorizer.transform(train_data)
    transformed_dev_data = vectorizer.transform(dev_data)
    
    lr_model = LogisticRegression(penalty='l2', C=100,tol=.01)
    lr_model.fit(transformed_train_data,train_labels)
    
    show_top10(lr_model, vectorizer, newsgroups_train.target_names)
    
    dev_predicted_labels = lr_model.predict(transformed_dev_data)
    print ('\n  The F1-Score for Predicting Dev Data for C=100 is {0:.3f} '.\
        format(metrics.f1_score(y_true=dev_labels, y_pred=dev_predicted_labels,average="macro")))
    
P8()
```

**ANSWER**

Instead of just making the most frequent words weight as 0 when using tf-idf, I am passing sub_linear=true, where it will add a value of 1 to the tf-idf log value and so the most repeated terms in the entire document corpus will get some weight. 

This improved the model accuracy by 0.5% on the dev data
